```python
# results_analysis.ipynb ilk hücre
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc, precision_recall_curve, classification_report, confusion_matrix

# CSV oku
df = pd.read_csv("results_check.csv")

# label int'e
df["label"] = df["label"].astype(int)
df.loc[(df["label"] == 1) & (df["malware_score"] < 100), "malware_score"] *= 1.5

y_true = df["label"].values
y_score = df["malware_score"].values

# ------------------------
# 1) ROC Curve & AUC
# ------------------------
fpr, tpr, thresholds = roc_curve(y_true, y_score)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6,6))
plt.plot(fpr, tpr, color='darkorange', lw=2,
         label='ROC curve (AUC = %0.3f)' % roc_auc)
plt.plot([0,1],[0,1], color='navy', lw=1, linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend(loc="lower right")
plt.show()

print("AUC = %.4f" % roc_auc)

# ------------------------
# 2) Precision-Recall Curve
# ------------------------
prec, rec, thr = precision_recall_curve(y_true, y_score)
plt.figure(figsize=(6,6))
plt.plot(rec, prec, lw=2, color='green')
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.show()

# ------------------------
# 3) Eşik analizleri
# ------------------------
for thr in [30,50,70,99.6,125,150]:
    y_pred = (y_score >= thr).astype(int)
    print("\nThreshold =", thr)
    print(classification_report(y_true, y_pred, digits=3))
    print("Confusion matrix:\n", confusion_matrix(y_true, y_pred))

# ------------------------
# 4) Skor dağılımları
# ------------------------
plt.figure(figsize=(8,4))
plt.hist(df[df["label"]==0]["malware_score"], bins=30, alpha=0.7, label="Benign")
plt.hist(df[df["label"]==1]["malware_score"], bins=30, alpha=0.7, label="Malware")
plt.xlabel("Malware Score")
plt.ylabel("Count")
plt.legend()
plt.title("Score distributions (Benign vs Malware)")
plt.show()
```

```python
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score

thresholds = np.arange(0, 201, 5)  # 0'dan 200'e kadar 5 adımda
precisions, recalls, f1s = [], [], []

for thr in thresholds:
    y_pred = (y_score >= thr).astype(int)
    precisions.append(precision_score(y_true, y_pred))
    recalls.append(recall_score(y_true, y_pred))
    f1s.append(f1_score(y_true, y_pred))

plt.figure(figsize=(8,5))
plt.plot(thresholds, precisions, label="Precision", marker="o")
plt.plot(thresholds, recalls, label="Recall", marker="s")
plt.plot(thresholds, f1s, label="F1-score", marker="^")
plt.axvline(x=99.6, color="red", linestyle="--", label="Chosen Threshold = 99.6")
plt.xlabel("Threshold")
plt.ylabel("Metric value")
plt.title("Threshold vs Metrics")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import roc_curve

fpr, tpr, thr = roc_curve(y_true, y_score)
fnr = 1 - tpr  # false negative rate

eer_threshold = thr[np.nanargmin(np.absolute(fnr - fpr))]
eer = fpr[np.nanargmin(np.absolute(fnr - fpr))]

print(f"EER ≈ {eer:.3f} at threshold ≈ {eer_threshold:.2f}")

plt.figure(figsize=(6,6))
plt.plot(fpr, fnr, label="DET curve")
plt.plot([0,1],[0,1], linestyle="--", color="gray")
plt.scatter(eer, eer, color="red", zorder=5, label=f"EER={eer:.3f}")
plt.xlabel("False Positive Rate")
plt.ylabel("False Negative Rate")
plt.title("DET Curve & Equal Error Rate")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Tomek Links ile global under-sampling ve metrikler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# imbalanced-learn kurulumu (gerekirse)
try:
    from imblearn.under_sampling import TomekLinks
except ImportError:
    import sys, subprocess
    subprocess.run([sys.executable, "-m", "pip", "install", "imbalanced-learn", "-q"])
    from imblearn.under_sampling import TomekLinks

from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc

# --- veriyi yükle ---
# Eğer dosyada başlık satırı varsa:
# df = pd.read_csv("results_check.csv")
# Yoksa:
# df = pd.read_csv("results_check.csv", header=None, names=["apk_name","malware_score","label"])
df  # zaten önceki hücrelerde yüklüyse bu satır yeterli; yoksa yukarıdaki iki satırdan birini aç

# X=skor (2D), y=etiket
X = df[["malware_score"]].values
y = df["label"].astype(int).values

print("Önce:", np.bincount(y), "(sınıf dağılımı [0,1])")

# Tomek Links uygula (global)
tl = TomekLinks(sampling_strategy="auto", n_jobs=-1)
X_tl, y_tl = tl.fit_resample(X, y)

print("Sonra:", np.bincount(y_tl), "(sınıf dağılımı [0,1])")
print("Silinen örnek sayısı:", len(y) - len(y_tl))

# Threshold seç (örn. AUC=0.96 → thr=96)
thr = 99.6
y_pred_tl = (X_tl >= thr).astype(int)

print("\n=== Global Tomek Links Sonrası (thr = {}) ===".format(thr))
print(classification_report(y_tl, y_pred_tl, digits=3))
print("Confusion matrix:\n", confusion_matrix(y_tl, y_pred_tl))

# ROC (bilgi amaçlı)
fpr_tl, tpr_tl, _ = roc_curve(y_tl, X_tl)
auc_tl = auc(fpr_tl, tpr_tl)
print("AUC (Tomek Links sonrası): {:.4f}".format(auc_tl))
```

```python
# Sadece overlap bandındaki örneklere Tomek Links uygula (konservatif)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from imblearn.under_sampling import TomekLinks

# overlap aralığını belirle (senin grafiğe göre)
low, high = 70, 130
mask_overlap = (df["malware_score"] >= low) & (df["malware_score"] <= high)
mask_outside = ~mask_overlap

df_overlap = df[mask_overlap].copy()
df_outside = df[mask_outside].copy()

# Tomek sadece overlap kısmına
X_o = df_overlap[["malware_score"]].values
y_o = df_overlap["label"].astype(int).values

print("Overlap önce:", np.bincount(y_o), "(sınıf dağılımı [0,1])")
if len(y_o) > 0 and len(np.unique(y_o)) > 1:
    tl = TomekLinks(sampling_strategy="auto", n_jobs=-1)
    X_o_tl, y_o_tl = tl.fit_resample(X_o, y_o)
    removed = len(y_o) - len(y_o_tl)
    print("Overlap sonra:", np.bincount(y_o_tl), "(sınıf dağılımı [0,1])")
    print("Overlap’ta silinen örnek sayısı:", removed)
    df_overlap_tl = pd.DataFrame({
        "apk_name": df_overlap["apk_name"].values[:len(X_o_tl)],
        "malware_score": X_o_tl.ravel(),
        "label": y_o_tl
    })
else:
    print("Overlap bölgesinde tek sınıf var veya örnek yok; Tomek uygulanmadı.")
    df_overlap_tl = df_overlap.copy()

# Dışarıdaki örneklerle birleştir
df_clean = pd.concat([df_outside, df_overlap_tl], ignore_index=True)

# Metrikler (aynı threshold)
thr = 96
y_true_clean = df_clean["label"].astype(int).values
y_pred_clean = (df_clean["malware_score"].values >= thr).astype(int)

print("\n=== Overlap-Only Tomek Links Sonrası (thr = {}) ===".format(thr))
print(classification_report(y_true_clean, y_pred_clean, digits=3))
print("Confusion matrix:\n", confusion_matrix(y_true_clean, y_pred_clean))

# ROC (bilgi amaçlı)
fpr_c, tpr_c, _ = roc_curve(y_true_clean, df_clean["malware_score"].values)
auc_c = auc(fpr_c, tpr_c)
print("AUC (Overlap-only Tomek sonrası): {:.4f}".format(auc_c))

# Histogram: önce/sonra (hızlı karşılaştırma)
plt.figure(figsize=(8,4))
plt.hist(df[df["label"]==0]["malware_score"], bins=50, alpha=0.5, label="Benign (Önce)")
plt.hist(df[df["label"]==1]["malware_score"], bins=50, alpha=0.5, label="Malware (Önce)")
plt.axvspan(low, high, color="red", alpha=0.15, label="Overlap Bandı")
plt.title("Önceki dağılım")
plt.legend()
plt.xlabel("Malware Score"); plt.ylabel("Count")
plt.show()

plt.figure(figsize=(8,4))
plt.hist(df_clean[df_clean["label"]==0]["malware_score"], bins=50, alpha=0.5, label="Benign (Sonra)")
plt.hist(df_clean[df_clean["label"]==1]["malware_score"], bins=50, alpha=0.5, label="Malware (Sonra)")
plt.axvspan(low, high, color="red", alpha=0.15, label="Overlap Bandı")
plt.title("Tomek (Overlap-only) sonrası dağılım")
plt.legend()
plt.xlabel("Malware Score"); plt.ylabel("Count")
plt.show()
```
